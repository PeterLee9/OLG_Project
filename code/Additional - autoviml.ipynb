```python
import pandas as pd
import numpy as np
from autoviml.Auto_ViML import Auto_ViML
from sklearn.model_selection import train_test_split
import warnings
```

```python
df_client = pd.read_csv("../data/raw/df_client.csv")
```

```python
df_client.loc[:,"Period"] = df_client.loc[:,"Period"].apply(lambda x: int(x[-1]))
```

```python
df_client = df_client.drop(columns=["BR Code", "Client", "Self_exclude_flag"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_client.drop(columns="risk_rating"), df_client["risk_rating"], test_size=0.2, random_state=0)
warnings.filterwarnings(action='ignore')
model, features, trainm, testm = Auto_ViML(
    X_train.join(y_train),
    "risk_rating",
    X_test.join(y_test),
    verbose=2,
)
```
