## 

```python
# correlation overview for all features
corr_client = df_client.iloc[:,6:].corr()
corr_client
```

```python
# correlation plot overview for all variable features
cmap = sns.diverging_palette(220, 10, as_cmap=True)
tick_labels = ["Variable_1"] + [""]*(corr_client.shape[1]-2) + ["Variable_30"]

fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot()
sns.heatmap(corr_client, ax=ax, cmap=cmap, annot=False, vmin=-1, vmax=1,
            xticklabels=tick_labels, yticklabels=tick_labels, square=True, cbar=False)
fig.colorbar(ax.get_children()[0], ax=ax, aspect=100, 
             location="top", orientation="horizontal")
ax.tick_params(axis='y', rotation=0)
plt.show()
```

```python
# time dependent correlation for each variable
cmap = sns.diverging_palette(220, 10, as_cmap=True)
fig, axs = plt.subplots(5, 6, figsize = (24,21))

for i in range(30):
    cols = list(filter(lambda x: x.startswith(f'Variable_{i+1}_'), df_client.columns))
    axs[i//6][i%6].set_title(f"Variable_{i+1}")
    sns.heatmap(data=df_client[cols].corr(), ax=axs[i//6][i%6], cmap=cmap, annot=True, vmin=-1, vmax=1,
                xticklabels=False, yticklabels=False, square=True, cbar=False)

fig.colorbar(axs[0][0].get_children()[0], ax=axs.ravel().tolist(), aspect=100, 
             location="top", orientation="horizontal", anchor=(0.5, 1.5))
fig.suptitle("Correlation between years (Y0, Y1, Y2, Y3) of each variable", fontsize=40)
fig.tight_layout()
fig.subplots_adjust(top=0.9)

plt.show()
```

```python
# correlation of variables in each year (between all Y0's, Y1's, etc.)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
fig, axs = plt.subplots(4, 1, figsize = (20,80))

for i in range(4):
    cols = list(filter(lambda x: x.endswith(f'_Y{i}'), df_client.columns))
    axs[i].set_title(f"Y{i}")
    sns.heatmap(data=df_client[cols].corr(), ax=axs[i], cmap=cmap, annot=True, vmin=-1, vmax=1,
                square=True, cbar=False)

fig.colorbar(axs[0].get_children()[0], ax=axs.ravel().tolist(), aspect=100, 
             location="top", orientation="horizontal", anchor=(0.5, 1.77))
fig.suptitle("Correlation between variables in each year", fontsize=40)
fig.tight_layout()
fig.subplots_adjust(top=0.958)

plt.show()
```
